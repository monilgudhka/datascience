Reading Titanic training data from csv file using pandas

```python
training_data = pd.read_csv('data/train.csv', index_col=0)
training_data[:10]
```

### Data Cleaning

Cleaning the data considering below points
- Name is not useful directly, but we can derive title from it
- Ticket will be unique for each passenger and the information it represents is already present in Cabin, Pclass and Embarked field
- Fare does not looks important as of now. (_need to check_)

```python
training_data['Title'] = training_data['Name'].apply(lambda x: x.split()[1].replace(',',''))
del training_data['Name']
del training_data['Ticket']
del training_data['Fare']
training_data[:10]
```
